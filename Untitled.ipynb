# preprocessing

```python


#definera columner
numeric_features = [
    "tau1", "tau2", "tau3", "tau4",
    "p1", "p2", "p3", "p4",
    "g1", "g2", "g3", "g4"
]


X = df[numeric_features]

# g4 √§r targeten
y = df["stabf"]  # categorical variable ('0' / '1')

#train / test split
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    random_state=42,
    stratify=y
)

#preprocessing pipeline
#byter ut saknade v√§rden med medianen
#standardisera parametrarna s√• att alla parametrar har samma medelv√§rde
numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])


# ColumnTransformer
# F√∂rhindrar data leakage
preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_features)
    ]
)


# Full pipeline
#kombinerar preprocessing och model tr√§ning
clf_baseline = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("model", LogisticRegression(max_iter=1000))
])

#tr√§na modellen
clf_baseline.fit(X_train, y_train)

y_pred = clf_baseline.predict(X_test)
#kolla hur accurate modellen √§r. 
baseline_accuracy = accuracy_score(y_test, y_pred)
baseline_accuracy
```

```python
print("Baseline accuracy:", baseline_accuracy)
print()
print("Classification report:")
print(classification_report(y_test, y_pred))
```
